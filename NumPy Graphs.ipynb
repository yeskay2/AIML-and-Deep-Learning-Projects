```python
# y = 5-3*x , x = [-5,5] , 20 points

x = np.linspace(-5,5,20)
y = 5-3*x
plt.plot(x,y)
plt.scatter(x,y)
plt.title("y = 5-3*x")
```

```python
#y=x^2 for x = -1,1
x = np.linspace(-1,1,50)
y = x**2
plt.plot(x,y)
# plt.scatter(x,y)
plt.title("$y = x^2$")
plt.axhline(color =  "black")
plt.axvline(color = "blue");
plt.grid(True) #color = "black")
```
